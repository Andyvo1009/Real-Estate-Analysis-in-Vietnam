```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

options = Options()
options.add_argument("--headless")  # Run in headless mode
#options.add_argument("--disable-gpu")  # Disable GPU (for compatibility)
options.add_argument("--no-sandbox")  # Avoid sandboxing issues
options.add_argument("--disable-blink-features=AutomationControlled")  # Reduce bot detection
options.add_argument("start-maximized")  # Maximize window
options.add_argument("window-size=1920,1080")  # Set window size
options.add_argument("--disable-blink-features=AutomationControlled")  # Reduce bot detection



# Spoof User-Agent to appear as a real user
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
)

driver=webdriver.Chrome(options=options)

id_list=[]

print(driver.title)
real_estate=[]

for c in range(1500):
  print(c)
  driver.get(f'https://batdongsan.com.vn/nha-dat-cho-thue/p{c+1}')
  time.sleep(1)
  
  
  for house in (driver.find_elements(By.CLASS_NAME,value="js__product-link-for-product-id"))[:-1]:
    house_info = {}
    product_id=house.get_attribute('data-product-id')
    print(product_id)
    try:
        address=house.find_element(By.CLASS_NAME,value='re__card-location').text
    except:
        address='None'
    try:
      price=house.find_element(By.CLASS_NAME,value='re__card-config-price').text
    except:
      price=0
    try:
      area=house.find_element(By.CLASS_NAME,value='re__card-config-area').text
    except:
      area=0
    time.sleep(0.1)
    try:
      bedrooms_num=house.find_element(By.CLASS_NAME,value='re__card-config-bedroom').text
    except Exception as e:
      bedrooms_num=0
    try:
      bathrooms_num=house.find_element(By.CLASS_NAME,value='re__card-config-toilet').text
    except:
      bathrooms_num=0
    house_info['product_id']=product_id
    house_info['address']=address
    house_info['price']=price
    house_info['area']=area
    house_info['bedrooms_num']=bedrooms_num
    house_info['bathrooms_num'] = bathrooms_num

    real_estate.append(house_info)

driver.quit()
df=pd.DataFrame(real_estate)

df.to_csv('rental_real_estate.csv',index=False)
df
```
