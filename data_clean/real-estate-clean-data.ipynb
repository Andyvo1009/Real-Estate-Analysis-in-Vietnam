```python
sale_df=pd.read_csv('/kaggle/input/real-estate-in-vietnam/sale_real_estate.csv')
sale_df
```

```python
sale_df['address']=sale_df['address'].str.replace('·\n','')
sale_df
```

```python
# Function to get the area
def get_area(row):
  area=row.split()[0]
  area=area.split('.')
  area=''.join(area)
  area=area.replace(',','.')

  return float(area)
sale_df['area']=sale_df['area'].apply(get_area)
sale_df
```

```python
# Function to get the price
def get_price(row):
  area=row['area']
  price=row['price'].split()[0].strip()
  product=row['price'].split()[1].strip()
  price=price.split('.')
  price=''.join(price)
  price=price.replace(',','.')
  if price=='Giá':
    price='deal'
  elif product=='tỷ':
    product=1e9
    price=float(price)*product
  elif product=='triệu':
    product=1e6
    price=float(price)*product
  elif product=='nghìn/m²':
    product=1e3
    price=float(price)*product
    price*=area
  return price
sale_df['price']=sale_df.apply(get_price,axis=1)
sale_df
```

```python
sale_df=sale_df[sale_df['price']!='deal']
sale_df
```

```python

def get_price(row):
  area=row['area']
  price=row['price'].split()[0].strip()
  product=row['price'].split()[1].strip()
  price=price.split('.')
  price=''.join(price)
  price=price.replace(',','.')
  if price=='Giá':
    price='deal'
  elif product=='triệu/tháng':
    product=1e6
    price=float(price)*product
  elif product=='nghìn/m²':
    product=1e3
    price=float(price)*product
    price*=area
  return price
rent_df['price']=rent_df.apply(get_price,axis=1)
rent_df
```
