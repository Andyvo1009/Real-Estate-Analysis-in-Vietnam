```python
sale_df=pd.read_csv('/kaggle/input/real-estate-in-vietnam/sale_real_estate.csv')
sale_df
```

```python
sale_df['address']=sale_df['address'].str.replace('·\n','')
sale_df
```

```python
# Function to get the area
def get_area(row):
  area=row.split()[0]
  area=area.split('.')
  area=''.join(area)
  area=area.replace(',','.')

  return float(area)
sale_df['area']=sale_df['area'].apply(get_area)
sale_df
```

```python
# Function to get the price
def get_price(row):
  area=row['area']
  price=row['price'].split()[0].strip()
  product=row['price'].split()[1].strip()
  price=price.split('.')
  price=''.join(price)
  price=price.replace(',','.')
  if price=='Giá':
    price='deal'
  elif product=='tỷ':
    product=1e9
    price=float(price)*product
  elif product=='triệu':
    product=1e6
    price=float(price)*product
  elif product=='nghìn/m²':
    product=1e3
    price=float(price)*product
    price*=area
  return price
sale_df['price']=sale_df.apply(get_price,axis=1)
sale_df
```

```python
sale_df=sale_df[sale_df['price']!='deal']
sale_df
```

# Create Model to Predict House Price in Ha Noi

```python
df=sale_df.query('city=="Hà Nội"')
df
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
df = df.drop(columns=['product_id', 'city', 'district'])

# Define numerical and categorical columns
numerical_cols = ['area', 'bedrooms_num']
categorical_cols = ['address']

# Define transformations
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),  # Standardize numerical columns
        ('cat', OneHotEncoder(), categorical_cols)  # One-hot encode categorical column
    ]
)

# Apply transformations
transformed_data = preprocessor.fit_transform(df)

# Convert sparse matrix to dense array (if necessary)
transformed_data_dense = transformed_data.toarray()
# transformed_data_dense = transformed_data.toarray()

# Retrieve feature names
num_scaled_cols = numerical_cols
cat_encoded_cols = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)
all_cols = num_scaled_cols + list(cat_encoded_cols)

# Create DataFrame
df_transformed = pd.DataFrame(transformed_data_dense, columns=all_cols)
df_transformed
```

```python

y=df['price']
X_train,X_test,y_train,y_test=train_test_split(df_transformed,y,random_state=42)
```

# After using GridSearchCV, the optimal parameters for Random Forest Regressor are as below


```python
# model to predict HN house price

rf=RandomForestRegressor(bootstrap=True,max_depth=30,
                         max_features='sqrt',
                         min_samples_leaf=1,
                         min_samples_split=10,
                         n_estimators=500   )
rf.fit(X_train,y_train)
```

```python
# R2 score
rf.score(X_test,y_test
        )
```
